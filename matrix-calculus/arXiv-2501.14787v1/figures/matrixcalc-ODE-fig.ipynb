```julia
using PyPlot
plt."rcParams"["font.family"] = "Times New Roman"
```

```julia
t = range(0, 2, length=1000)
```

```julia
plot(t, exp.(-t), "r-", linewidth=2)
plot(t, exp.(-(1 + 0.1)*t), "b--", linewidth=2)
xlabel(L"“time” $t$", math_fontfamily="dejavuserif")
ylabel(L"solution $u(p, t)$", math_fontfamily="dejavuserif")
title(L"solutions to $\frac{\partial u}{\partial t} = -p u$ with $u(p,0) = 1$", math_fontfamily="dejavuserif", y=0.9, fontsize=16)
text(0.5, exp(-0.5)-0.09, L"u(p, t)", math_fontfamily="dejavuserif", fontsize=16, color="red", rotation=-39)
text(0.75, exp(-0.75)-0.3, L"u(p + \delta p, t)", math_fontfamily="dejavuserif", fontsize=16, color="blue", rotation=-31)

arrow(1.5, exp(-(1 + 0.1)*1.5) - 0.1, 0, 0.1, width=0.01, length_includes_head=true, color="black")
arrow(1.5, exp(-1 * 1.5) + 0.1, 0, -0.1, width=0.01, length_includes_head=true, color="black")
text(1.55, exp(-1 * 1.5) + 0.035, L"\delta u", math_fontfamily="dejavuserif", fontsize=16)

ylim(0, 1)
xlim(0,2)
xticks([0, 0.5, 1.0, 1.5, 2.0])
savefig("ode-fig.pdf")
```
